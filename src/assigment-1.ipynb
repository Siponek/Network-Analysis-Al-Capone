### Plotting Functions

```python
def calculate_log_bins(degrees):
    """Calculate log bin centers and counts for trend line plotting."""
    # Let NumPy automatically determine the bin size
    hist, edges = np.histogram(degrees, bins='auto')
    non_zero = hist > 0
    log_bin_centers = np.log(0.5 * (edges[:-1] + edges[1:]))[non_zero]
    log_hist = np.log(hist)[non_zero]
    return log_bin_centers, log_hist, edges

def plot_degree_distribution(degrees, bins : str|range= 'auto', name: str = "degree_distribution.png"):
    """Plot the degree distribution on a normal scale."""
    plt.figure(figsize=(10, 6))
    sns.histplot(degrees, bins=bins, alpha=0.5,color='blue', edgecolor='gray', linewidth=1.2, kde=True)
    plt.title("Degree Distribution")
    plt.xlabel("Degree")
    plt.ylabel("Number of Nodes")
    plt.savefig(out_path.joinpath("img/" + name))
    plt.show()

def plot_log_log_scale(degrees, log_bin_centers, log_hist, bins= 'auto', name: str = "degree_distribution_log_log.png"):
    """Plot the degree distribution and trend line on a log scale."""
    # Perform linear regression on log-transformed data
    slope, intercept, _, _, _ = linregress(log_bin_centers, log_hist)
    plt.figure(figsize=(10, 6))
    sns.histplot(degrees, bins=bins, alpha=0.5, color='blue', edgecolor='gray', linewidth=1.2, kde=True, log_scale=True)
    plt.yscale('log')
    # Plot trend line
    plt.plot(np.exp(log_bin_centers), np.exp(intercept + slope * log_bin_centers), 'r-', label=f'Trend line: slope={slope:.2f}')
    
    plt.title("Degree Distribution on Log-Log Scale")
    plt.xlabel("Degree")
    plt.ylabel("Number of Nodes (log scale)")
    plt.legend()
    plt.savefig(out_path.joinpath("img/" + name))
    plt.show()
```

### Degree Distribution

```python
sns.set_style("whitegrid")
sns.despine()
# sns.set_theme("paper", font_scale=1.5)
degrees = [degree for _, degree in G.degree()]
bins = range(0, max(degrees), 1)
log_bin_centers, log_hist, _ = calculate_log_bins(degrees)
plot_degree_distribution(degrees, bins)
plot_log_log_scale(degrees, log_bin_centers, log_hist)
```

```python
# Log-transform degree and counts for linear regression
hist, edges = np.histogram(degrees, bins=bins)
non_zero = hist > 0
log_bin_centers = np.log(0.5 * (edges[:-1] + edges[1:]))[non_zero]
log_hist = np.log(hist)[non_zero]

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(log_bin_centers, log_hist)

# Plot original histogram
plt.hist(degrees, bins=bins, alpha=0.5, color='b', edgecolor='black', linewidth=1.2, log=True)
plt.xscale('log')
plt.yscale('log')

# Plot trend line
plt.plot(np.exp(log_bin_centers), np.exp(intercept + slope*log_bin_centers), 'r-', label=f"Trend line: slope={slope:.2f}")

plt.title("Degree Distribution with Trend Line on Log-Log Scale")
plt.xlabel("Degree")
plt.ylabel("Number of Nodes")
plt.legend()
plt.show()
```

### Community detection

#### Greedy modulatiry optimization

```python
communities = nx.algorithms.community.greedy_modularity_communities(G)
amount_of_communities = len(communities)


# Create a dictionary that maps each node to its community, because the community detection algorithm returns a list of sets
community_map = {node: i for i, community in enumerate(communities) for node in community} # type: ignore
# Get the largest community
largest_community = max(communities, key=len)
# Calculate modularity and coverage
communities_list = list(nx.algorithms.community.greedy_modularity_communities(G))

modularity = nx.algorithms.community.quality.modularity(G, [*communities_list])
coverage = nx.algorithms.community.quality.partition_quality(G, [*communities_list])

pos = nx.spring_layout(G)
cmap = plt.cm.get_cmap('viridis', amount_of_communities)

# Draw nodes with colors assigned based on their community
nx.draw_networkx_nodes(G, pos, node_size=40, cmap=cmap, 
                       node_color=[community_map[node] for node in G.nodes()])  # type: ignore

nx.draw_networkx_edges(G, pos, alpha=0.5)
print(f"Number of communities: {amount_of_communities}")
print("Modularity of the largest community: ", modularity)
print("Coverage of the largest community: ", coverage)
plt.show()
```

#### Random network comparison

```python
random_graph = nx.gnm_random_graph(G.number_of_nodes(), G.number_of_edges())
print("Average clustering coefficient of the original network:", nx.average_clustering(G))
print("Average clustering coefficient of a comparable random network:", nx.average_clustering(random_graph))

# Calculate degree distribution of random graph
random_degrees = [degree for _, degree in random_graph.degree()]
bins = range(0, max(random_degrees), 1)
log_bin_centers, log_hist, _ = calculate_log_bins(random_degrees)
plot_degree_distribution(random_degrees, bins, name="random_degree_distribution.png")
plot_log_log_scale(random_degrees, log_bin_centers, log_hist, name = "random_degree_distribution_log_log.png")
# Calculate avg path length of random graph
if nx.is_connected(random_graph):
    avg_path_length = nx.average_shortest_path_length(random_graph)
    print(f"Average path length of random graph: {avg_path_length}")
else:
    largest_cc = max(nx.connected_components(random_graph), key=len)
    subgraph = random_graph.subgraph(largest_cc)
    print(f"Average clustering coefficient of the LCC: {nx.average_clustering(subgraph)}")
# Calculate connectivity and robustness of random graph
connected_components_count = len(list(nx.connected_components(G)))
random_connected_components_count = len(list(nx.connected_components(random_graph)))
print(f"Number of connected components random to mouse network: {random_connected_components_count} to {connected_components_count}")
avg_clustering_coefficient = nx.average_clustering(random_graph)
print(f"Average Clustering Coefficient: {avg_clustering_coefficient}")

```
